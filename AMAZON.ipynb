```python
# Checking for the presence of outliers in the dataset using the IQR method
def detect_outliers_iqr(data):
    outliers_dict = {} # Setting the outlier dictionary to store the outliers.
    for column in data.select_dtypes(include =['number']).columns:
            Q1 = data[column].quantile(0.25)
            Q3 = data[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 -1.5*IQR
            upper_bound = Q3 - 1.5*IQR
            #Identifying the outliers
            outliers = data[(data[column]<lower_bound)|(data[column]>upper_bound)]
            outliers_dict[column] = len(outliers)
    return outliers_dict
print(detect_outliers_iqr(Amazon))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Function to plot all boxplots in one chart with subplots
def plot_boxplots_in_grid(data):
    numeric_columns = data.select_dtypes(include=['number']).columns
    num_columns = len(numeric_columns)
    
    # Create a grid of subplots (determine rows and columns based on number of numeric columns)
    cols = 2  
    rows = (num_columns // cols) + (num_columns % cols > 0)  
    
    # Create a figure with subplots
    fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15, 6 * rows))
    axes = axes.flatten() 
    
    # Plot each box plot in a separate subplot
    for i, column in enumerate(numeric_columns):
        sns.boxplot(x=data[column], ax=axes[i])
        axes[i].set_title(f"Box Plot of {column}")
        axes[i].set_xlabel("Values")
    
    # Remove any empty subplots if columns do not fill the grid completely
    for j in range(i+1, len(axes)):
        fig.delaxes(axes[j])
    
    # Adjust layout
    plt.tight_layout()
    plt.show()
```

```python
def remove_outliers_iqr(data):
    # Make a copy of the original dataset to avoid modifying it directly
    cleaned_data = data.copy()

    # Loop through numeric columns and remove rows with outliers
    for column in cleaned_data.select_dtypes(include=['number']).columns:
        Q1 = cleaned_data[column].quantile(0.25)
        Q3 = cleaned_data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Remove rows where the value is below the lower bound or above the upper bound
        cleaned_data = cleaned_data[(cleaned_data[column] >= lower_bound) & (cleaned_data[column] <= upper_bound)]

    return cleaned_data

# Apply the function to clean the dataset
Amazon_cleaned = remove_outliers_iqr(Amazon)

# Print the number of rows after removing outliers
print("Data after removing outliers:", Amazon_cleaned.shape)
```

```python
def cap_outliers_iqr(data):
    # Make a copy of the original dataset to avoid modifying it directly
    capped_data = data.copy()

    # Loop through numeric columns and cap outliers
    for column in capped_data.select_dtypes(include=['number']).columns:
        Q1 = capped_data[column].quantile(0.25)
        Q3 = capped_data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Cap values below the lower bound and above the upper bound
        capped_data[column] = capped_data[column].apply(lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x)

    return capped_data

# Apply the function to cap the outliers
Amazon_capped = cap_outliers_iqr(Amazon)

# Print the number of rows after capping outliers
print("Data after capping outliers:", Amazon_capped.shape)
```

##### Using Apriori Algorithm

```python
# Encoding the data using one- hot encoder since the data is in categorical form
from mlxtend.preprocessing import TransactionEncoder

# Step 1: Prepare transactions by grouping products by user_id
transactions = Amazon.groupby('user_id')['product_name'].apply(list).reset_index()

# Step 2: Use TransactionEncoder to one-hot encode the transactions
te = TransactionEncoder()
te_ary = te.fit(transactions['product_name']).transform(transactions['product_name'])

# Convert the encoded transactions to a DataFrame
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

# Display the first few rows of the encoded DataFrame to verify
print(df_encoded.head())

```

```python
print(df_encoded.tail())  # Check the top 5 rows
print(df_encoded.sum())   # Check if there are any products with non-zero values
```

```python
# Remove products that appear in very few transactions
df_encoded = df_encoded.loc[:, (df_encoded.sum() > 1)]  # Keep products appearing more than once

# Rerun Apriori
frequent_itemsets = apriori(df_encoded, min_support=0.001, use_colnames=True)
print(frequent_itemsets.head())
```

```python
print(frequent_itemsets.head())  # Check the first few itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

# Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

# Display the first few rules
print(rules.head())
```

```python
import matplotlib.pyplot as plt

plt.scatter(rules['confidence'], rules['lift'], alpha=0.5)
plt.xlabel('Confidence')
plt.ylabel('Lift')
plt.title('Association Rules: Confidence vs. Lift')
plt.show()
```


##### Items on the left-hand side of the rule (the if-condition).
##### Example: pTron Solero TB301 3A Type-C Data and Fast Charging Cable
##### Consequents:

##### Items on the right-hand side of the rule (the then-outcome).
##### Example: pTron Solero MB301 3A Micro USB Data & Charging Cable
##### Antecedent Support:

##### The proportion of transactions containing the antecedent.
##### Example: 0.001675 means 0.1675% of transactions included pTron Solero TB301 3A Cable.
##### Consequent Support:

##### The proportion of transactions containing the consequent (similar to antecedent support).
##### Example: 0.001675 means the consequent appeared in 0.1675% of transactions.
##### Support:

##### The proportion of transactions containing both the antecedent and consequent.
##### Example: 0.001675 means that both items were present together in 0.1675% of all transactions.
##### Confidence:

##### Probability that a transaction containing the antecedent will also contain the consequent.
##### Example: 1.0 means 100% confidence – every time TB301 Type-C Cable was bought, the MB301 Micro USB Cable was also bought.
##### Lift:

##### How much more likely the consequent is to occur compared to its expected frequency, assuming the antecedent occurs.
##### Example: Lift = 597 means that the two products are 597 times more likely to appear together than by chance, indicating a strong association.
##### Leverage:

##### The difference between the observed support and expected support, if the antecedent and consequent were independent.
##### Example: A positive value like 0.001672 suggests a useful association.
##### Conviction:

##### A measure of how strongly the rule implies the consequent. If conviction is infinity (inf), it means the consequent always follows the antecedent.
##### Zhang's Metric:

##### A metric that evaluates the reliability of the rule. A value of 1.0 indicates a perfect association between the antecedent and the consequent.


##### Rule 1:
##### If a user buys pTron Solero TB301 Type-C Cable, they always buy pTron Solero MB301 Micro USB Cable (100% confidence, lift = 597).
##### This is a strong association, indicating the two products are often bought together, possibly as a bundle or complementary items.
##### Rule 2:
##### The reverse of the first rule: If a user buys the MB301 Micro USB Cable, they always buy the TB301 Type-C Cable as well.


##### Using FP- Growth Algorithm

```python
from mlxtend.frequent_patterns import fpgrowth, association_rules

# Step 1: Apply FP-Growth with a lower min_support
frequent_itemsets_fp = fpgrowth(df_encoded, min_support=0.001, use_colnames=True)

# Step 2: Display the frequent itemsets
print("Frequent Itemsets from FP-Growth:")
print(frequent_itemsets_fp.head())

# Step 3: Generate association rules from the frequent itemsets
rules_fp = association_rules(frequent_itemsets_fp, metric="confidence", min_threshold=0.5)

# Step 4: Display the first few rules
print("Association Rules from FP-Growth:")
print(rules_fp.head())
```

##### Interpretation of the Rules
##### Rule 1:
##### If a user buys the pTron Solero TB301 3A Type-C Cable, they always buy the MB301 Micro USB Cable.
#####  Lift = 597.0 means that the occurrence of the two products together is 597 times higher than expected by random chance. This indicates a very strong relationship between these products.
#####  Rule 2: The reverse rule shows that every user who buys the Micro USB Cable also buys the Type-C Cable.

#####  Context-Based ARM

```python
# Example: Add a 'context' column to simulate a discount context
Amazon['context'] = Amazon['discount_percentage'].apply(lambda x: 'High Discount' if x > 50 else 'Low Discount')

# Group transactions by user_id and context to build contextual transactions
transactions_contextual = Amazon.groupby(['user_id', 'context'])['product_name'].apply(list).reset_index()

print(transactions_contextual.head())  # Verify the structure
```

```python
from mlxtend.preprocessing import TransactionEncoder

# Encode the contextual transactions into one-hot format
te = TransactionEncoder()
te_ary = te.fit(transactions_contextual['product_name']).transform(transactions_contextual['product_name'])

df_encoded_contextual = pd.DataFrame(te_ary, columns=te.columns_)

print(df_encoded_contextual.head())  # Verify the one-hot encoding
```

```python
from mlxtend.frequent_patterns import fpgrowth, association_rules

# Run FP-Growth on the contextual one-hot encoded data
frequent_itemsets_context = fpgrowth(df_encoded_contextual, min_support=0.01, use_colnames=True)

print("Frequent Itemsets with Context:")
print(frequent_itemsets_context.head())
```
